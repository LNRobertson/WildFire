```python
df['DISCOVERY_DATE'] = pd.to_datetime(df['DISCOVERY_DATE'])
df['CONT_DATE'] = pd.to_datetime(df['CONT_DATE'])
```

```python
# Sort the data by date
df = df.sort_values(by='DISCOVERY_DATE')
```

```python
# Group the data by year and count the number of fires
fires_per_year = df.groupby(df['DISCOVERY_DATE'].dt.year).size()

# Plot the number of fires per year
fires_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Fires')
plt.show()
```

```python
# Group the data by year and sum the fire size
fire_size_per_year = df.groupby(df['DISCOVERY_DATE'].dt.year)['FIRE_SIZE'].sum()

# Plot the total fire size per year
fire_size_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Fire Size (acres)')
plt.show()
```

```python
df.DISCOVERY_DATE.info()
```

```python
# Group the data by year and cause, and calculate the total fire size for each group
df_grouped = df.groupby(['FIRE_YEAR', 'NWCG_GENERAL_CAUSE'])['FIRE_SIZE'].sum().unstack()

# Create a stacked bar chart
df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the chart title and labels
plt.title('Total Fire Size by Year and Cause')
plt.xlabel('Year')
plt.ylabel('Total Fire Size')

# Show the chart
plt.show()
```
