```python
df['DISCOVERY_DATE'] = pd.to_datetime(df['DISCOVERY_DATE'], errors='coerce', format='%Y-%m-%d')
df['CONT_DATE'] = pd.to_datetime(df['CONT_DATE'], errors='coerce', format='%Y-%m-%d')
```

```python
# Remove any rows with missing values in the DISCOVERY_DATE or CONT_DATE columns
df = df.dropna(subset=['DISCOVERY_DATE', 'CONT_DATE'])
```

```python
# Sort the data by date
df = df.sort_values(by='DISCOVERY_DATE')
```

```python
# Group the data by year and count the number of fires
fires_per_year = df.groupby(df['DISCOVERY_DATE'].dt.year).size()

# Plot the number of fires per year
fires_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Fires')
plt.show()
```

```python
# Group the data by year and sum the fire size
fire_size_per_year = df.groupby(df['DISCOVERY_DATE'].dt.year)['FIRE_SIZE'].sum()

# Plot the total fire size per year
fire_size_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Fire Size (acres)')
plt.show()
```
